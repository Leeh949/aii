````python
plt.figure(figsize=(10,3))
# 기본 미적 속성 설정
sns.set(style='darkgrid', context='notebook',palette='Set1', rc={'figure.figsize':(10,3)})

plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False)


ax= sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species',palette='hsv')
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1,1))
ax.set(title='상관관계가 높은 두 변수', ylabel="꽃잎너비")
ax.set_xlabel("꽃잎 길이",fontsize=20)
ax.set_xticks([1,3,5,7])
plt.show()
````

## 2.3 seaborn 함수로 그래프 그리기 
### 1) 관계형 그래프
- scatterplot, lineplot, relplot(서브플롯 제공): x가 수치형

````python
# 종별 다른색상과 다른 스타일로 scatterplot
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species',style='species',palette='Set1')
````

````python
# 종별 lineplot으로 출력
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species',style='species',palette='Set1')
````

````python
# 종별 scatterplot을 서브플롯으로 
 
# kind: 서브플롯으로 그려질 그래프 종류, 
# col: col단위로 서브플롯 표현, 
#한줄에 표현될 서브플롯 수 # col_wrap=2
# facet_kws={'sharey':False} # 축 공유 
g = sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species',
            kind='line',col='species',col_wrap=2)

#크기 조정
g.fig.set_size_inches(10,3)

````

````python
# ax [0]에는 scatter, ax[1]에는 line
fig, ax= plt.subplots(ncols=2)
sns.scatterplot(data=iris, x= 'petal_length',y='petal_width',hue='species', ax= ax[0])
sns.lineplot(data=iris, x= 'petal_length',y='petal_width',hue='species', ax= ax[1])
````

### 2) 범주형 그래프
- stripplot(무작위로 흩뿌림), warmplot(겹치지않음) - 점도표
- barplot, barhplot, pointplot(막대대신 marker가 있는 선)
- countplot(bar는 y의 평균값이 높이, count는 x의 빈도수 높이)
- boxplot, violinplot, boxenplot        => catplot

````python
#x축이 범주형일 때의 산점도
sns.stripplot(data=iris, x='species', y='petal_width', hue='species',
             palette='Set1')
````

````python
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species',
             palette='Set1')
````

````python
# 내림차순으로 정렬된 iris 범주형 산점도
sns.swarmplot(data=iris.sort_values('petal_width',ascending=False),
             x='species', y='petal_width', hue='species',
             palette='Set1')
````

````python
# x축의 순서를 임의로 정할 수 있음
sns.swarmplot(data=iris,
             x='species', y='petal_width', hue='species', palette='Set1',
             order=['versicolor','setosa','virginica'])
plt.show()
````

````python
# barplot: 각 범주별 평균값과 그에 대한 신뢰구간(신뢰구간제외: errorbar=None)
sns.barplot(data=iris, x='species', y='sepal_width', errorbar=None)
sns.lineplot(data=iris, x='species',y='sepal_width',color='k')
````

````python
sns.pointplot(data=iris, x='species', y='sepal_width')
````
